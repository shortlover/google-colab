```python
import numpy as np
import matplotlib.pyplot as plt #Fixed typo here
from scipy.stats import norm
```

```python
r = 0.00
sig = 0.2
T = 30/365
M = 1000
N = 300

dt = T/N
rdt = r*dt   # r X dt
sigsdt = sig* np.sqrt(dt)   # sigma X   square root
```

```python
S0 = 100
np.random.seed(100)
S = np.empty([M,N+1])
rv = np.random.normal(r*dt,sigsdt,[M,N] )
```

```python
# for i in range(M):
#     S[i,0] = S0
#     for j in range(N):
#         S[i,j+1] = S[i,j] * (1+np.random.normal(r*dt,sigsdt))
```

```python
for i in range(M):
    S[i,0] = S0
    for j in range(N):
        S[i,j+1] = S[i,j] * (1+rv[i,j])


for i in range(M):
    plt.plot(S[i,:])

plt.show()
```

```python
M = 64
K = 100

hedge = 0
cost = 0
for j in range(N):
    d1 = (np.log(S[M,j]/K)+(r+0.5*sig**2)*(T-j*dt))/(sig*np.sqrt(T-j*dt))
    delta = norm.cdf(d1)
    cost = cost + (hedge-delta) * S[M,j]
    hedge = delta
    #print(S[M,j], delta, cost)

if S[M,N] > K:
    delta = 1
    cost = cost + K
else:
    delta = 0
    cost = cost + S[M,N] * hedge

print(S[M,N], 0, cost)
```
